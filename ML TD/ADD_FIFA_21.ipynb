### Linear Regression Graph

Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data. The graph typically shows data points and a straight line that best describes the relationship between these points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data for linear regression
X_linear = np.random.rand(100, 1) * 10
y_linear = 2 * X_linear + 1 + np.random.randn(100, 1) * 2

# Fit a linear regression model
model_linear = LinearRegression()
model_linear.fit(X_linear, y_linear)
y_linear_pred = model_linear.predict(X_linear)

# Plotting the linear regression graph
plt.figure(figsize=(8, 6))
plt.scatter(X_linear, y_linear, label='Actual Data')
plt.plot(X_linear, y_linear_pred, color='red', label='Linear Regression Line')
plt.title('Linear Regression Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

### Logistic Regression Graph

Logistic regression is used for binary classification problems. It models the probability of a certain class or event occurring, typically producing an S-shaped (sigmoid) curve. The output is a probability value between 0 and 1.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Generate synthetic data for logistic regression
X_logistic = np.random.rand(100, 1) * 10 - 5 # Centered around 0
y_logistic = (1 / (1 + np.exp(-(2 * X_logistic + np.random.randn(100, 1) * 0.5))) > 0.5).astype(int)

# Fit a logistic regression model (using a simple solver for demonstration)
model_logistic = LogisticRegression(solver='liblinear')
model_logistic.fit(X_logistic, y_logistic.ravel())

# Generate a smooth curve for plotting the logistic function
X_test = np.linspace(-5, 5, 300).reshape(-1, 1)
prob_pred = model_logistic.predict_proba(X_test)[:, 1]

# Plotting the logistic regression graph
plt.figure(figsize=(8, 6))
plt.scatter(X_logistic, y_logistic, label='Actual Data (0 or 1)', alpha=0.6)
plt.plot(X_test, prob_pred, color='red', label='Logistic Regression Curve')
plt.title('Logistic Regression Example')
plt.xlabel('X-axis')
plt.ylabel('Probability of Y=1')
plt.legend()
plt.grid(True)
plt.show()
```
